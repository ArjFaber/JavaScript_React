```python
import pandas as pd
import numpy as np

# Read the Excel file into a DataFrame
df = pd.read_excel('./Data/finalInventory.xlsx', engine='openpyxl')

colNamesJSON = ["Desc", "SAP_ID", "numItems", "Palette","Date_de_lachat", "Montant_de_lachat", "ValeurTot", "Remark", "vendName"]

colNames = np.asarray(df.columns)
data = pd.DataFrame({'job_ID':list(range(0, len(df)))})

for j in range(0,len(colNamesJSON)):
    data[colNamesJSON[j]] = df[colNames[j]].astype(str)


```

```python
data = data.transpose()
```
